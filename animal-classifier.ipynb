```python
 import  tensorflow as tf
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
```

```python
filename = "C:/Users/SPECTRE/Desktop/PES/Semester 3/Mini Project/Shero.jpeg"
```

```python
img = image.load_img(filename, target_size = (224,224))
```

```python
plt.imshow(img)
```

```python
mobile = tf.keras.applications.resnet50.ResNet50()
```

```python
from tensorflow.keras.applications import imagenet_utils
```

```python
resized_img = image.img_to_array(img) 
final_image = np.expand_dims(resized_img,axis=0)  ##Fourth Dimension Image
final_image= tf.keras.applications.resnet50.preprocess_input(final_image)
```

```python
final_image.shape
```

```python
predictions = mobile.predict(final_image)
```

```python
results = imagenet_utils.decode_predictions(predictions)
```
